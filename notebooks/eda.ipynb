# Medical QA System - Dataset EDA
## Exploring PubMedQA Dataset

```python
# Check versions of key libraries
import torch
import transformers
import datasets

print(f"PyTorch version: {torch.__version__}")
print(f"Transformers version: {transformers.__version__}")
print(f"Datasets version: {datasets.__version__}")
```

```python
# Load the PubMedQA dataset in labeled format
from datasets import load_dataset

dataset = load_dataset("pubmed_qa", "pqa_labeled")
print(dataset)
```

```python
# Explore the structure by showing the first example from the train split
first_example = dataset["train"][0]
print("First Example:")
print(f"pubid: {first_example['pubid']}")
print(f"question: {first_example['question']}")
print(f"context: {first_example['context']}")
print(f"long_answer: {first_example['long_answer']}")
print(f"final_decision: {first_example['final_decision']}")
```

```python
# Print basic statistics about the dataset
train_split = dataset["train"]
print(f"Total number of examples in train split: {len(train_split)}")
print(f"\nUnique values in 'final_decision' field:")

final_decisions = train_split["final_decision"]
from collections import Counter
decision_counts = Counter(final_decisions)
for decision, count in sorted(decision_counts.items()):
    print(f"  {decision}: {count}")
```

```python
# Analyze the length of questions in the dataset
questions = train_split["question"]
question_lengths = [len(q.split()) for q in questions]

print("Question Length Analysis:")
print(f"  Minimum length: {min(question_lengths)} words")
print(f"  Maximum length: {max(question_lengths)} words")
print(f"  Average length: {sum(question_lengths) / len(question_lengths):.2f} words")
```

```python
# Analyze the length of long_answer field in the dataset
answers = train_split["long_answer"]
answer_lengths = [len(a.split()) for a in answers]

print("Long Answer Length Analysis:")
print(f"  Minimum length: {min(answer_lengths)} words")
print(f"  Maximum length: {max(answer_lengths)} words")
print(f"  Average length: {sum(answer_lengths) / len(answer_lengths):.2f} words")
```

```python
# Display 3 random examples showing question, final_decision, and first 200 chars of long_answer
import random

random_indices = random.sample(range(len(train_split)), 3)

for i, idx in enumerate(random_indices, 1):
    example = train_split[idx]
    print(f"\n--- Sample {i} ---")
    print(f"Question: {example['question']}")
    print(f"Final Decision: {example['final_decision']}")
    print(f"Answer (first 200 chars): {example['long_answer'][:200]}...")
```
